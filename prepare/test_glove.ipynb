```python
for ptfile in files:
    print("processing:", os.path.join(filepathR, ptfile))
    tmpfile = torch.load(os.path.join(filepathR, ptfile))
    features = list()
    for frame_id in range(tmpfile.size()[0]):
        feature = torch.zeros((1, 300))
        feature_num = 0
        frame = tmpfile[frame_id, :]
        frame_non = torch.nonzero(frame)
        for item in frame_non[:]:
            feature += embedding_name[item.item()] * frame[item].item()
            feature_num += frame[item].item()
            
        feature = feature / feature_num
        features.append(feature)
        
    feature_file = torch.cat(features, 0)
    torch.save(feature_file, os.path.join(filepathW, ptfile))
    assert tmpfile.size(0) == feature_file.size(0)
    print("saved features in", os.path.join(filepathW, ptfile))
```

```python
import torch
import pickle
from nltk.corpus import wordnet as wn
from nltk.stem import WordNetLemmatizer
from flair.data import Sentence
from flair.embeddings import TransformerWordEmbeddings


def generate_embeddings_synsets(save_embedding_file, embedding=TransformerWordEmbeddings()):
    """
    :param save_embedding_file: file name to save the hash map of synsets' names mapped to their respective embeddings
    :param embedding: flair.embeddings (e.g. BertEmbeddings())
    :return: None
    """
    syn_embeddings = {}
    lemmatizer = WordNetLemmatizer()
    for i, syn in enumerate(list(sorted(wn.all_synsets(), key=lambda syn: syn.name()))):
        if not syn.examples():
            continue

        syn_name = syn.name().split('.')[0]
        syn_pos = syn.name().split('.')[1]
        print(i, syn.name())
        embed = torch.zeros((768))
        num_embed = 0
        for sent in syn.examples():
            sentence = Sentence(sent.lower(), use_tokenizer=True)
            embedding.embed(sentence)
            for token in sentence.tokens:
                try:
                    lemmatized_word = lemmatizer.lemmatize(token.text, syn_pos)
                    if lemmatized_word == syn_name:
                        embed.add_(token.embedding.cpu())
                        num_embed += 1
                except Exception as e:
                    print (e)
                    print("Error with token:", token.text, "when processing", syn.name())

        if any(embed):
            # non-zero tensor
            syn_embeddings[syn.name()] = embed / num_embed
    pickle.dump(syn_embeddings, open(save_embedding_file, 'wb'))
```

```python
generate_embeddings_synsets('../data/synset.pkl')
# import nltk
# nltk.download('wordnet')
```
